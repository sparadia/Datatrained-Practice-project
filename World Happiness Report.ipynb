# Introduction

The World Happiness Report is a seminal study of happiness around the world. The first report was released in 2012, followed by the second in 2013, the third in 2015, and finally the fourth in 2016. The World Happiness Report 2017, which rates 155 countries based on their happiness levels, was released on March 20th during a United Nations ceremony commemorating International Day of Happiness. Governments, companies, and civil society are increasingly using happiness indices to influence policy-making decisions, and the report is gaining global reputation. Leading experts from a variety of sectors – economics, psychology, survey analysis, national statistics, health, public policy, and more – explain how well-being assessments can be used to analyse a country's progress. The reports examine the level of happiness in the United States.

# Content

The Gallup World Poll data was used to calculate the happiness scores and rankings. The results are based on responses to the poll's main life appraisal question. This question, known as the Cantril ladder, asks respondents to imagine a ladder with a 10 representing the best conceivable life and a 0 representing the worst possible life, and to rank their existing lives on that scale. The results are based on nationally representative samples from 2013 to 2016 and are weighted using Gallup weights to ensure that the estimates are representative. The columns that follow the happiness score assess how much each of six characteristics - economic production, social support, life expectancy, freedom, lack of corruption, and generosity – contribute to making life evaluations greater in each of the six dimensions.

# What is Dystopia?

Dystopia is a fictional country populated by the world's unhappiest people. The goal of creating Dystopia was to create a baseline against which all countries could be compared favourably (no country performs worse than Dystopia) in terms of each of the six major criteria, allowing each sub-bar to have a positive width. Dystopia is defined by the lowest scores observed for the six main variables. In contrast to Utopia, life would be extremely unpleasant in a country with the world's lowest earnings, lowest life expectancy, lowest generosity, greatest corruption, least freedom, and least social assistance.

# What is residual?

The residuals, or unexplained components, vary by country, indicating how much the six variables over- or under-explain average 2014-2016 life ratings. Over the whole set of countries, these residuals have an average value of about zero. When the equation in Table 2.1 is applied to average 2014-2016 data for the six variables in a country, the average residual is shown in Figure 2.2. These residuals are coupled with the estimate for life assessments in Dystopia, resulting in a composite bar with always positive values. Although certain life evaluation residuals are fairly significant, occasionally exceeding one point on a scale of 0 to 10, as shown in Figure 2.2, they are always far smaller than the estimated residuals.
